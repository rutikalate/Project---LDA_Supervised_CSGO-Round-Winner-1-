```python
# Converting categorical features into a integer column
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
df['bomb_planted'] = le.fit_transform(df['bomb_planted'])
df['map'] = le.fit_transform(df['map'])
df['round_winner'] = le.fit_transform(df['round_winner'])
```

```python
X = df.drop(columns=["round_winner"])
y = df[["round_winner"]]
```

```python
X
```

```python
y
```

```python
#df.duplicated()
#drop_duplicates
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=999)
```

```python
x_train
```

```python
# Scaling the data
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
# Applying Linear Discriminant Analysis
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda = LinearDiscriminantAnalysis()
lda.fit(x_train, y_train) #min(2-1, 96) ==> 1
```

```python
x_test
```

```python
lda.transform(x_test)
```

```python
lda.coef_
```

```python
# Obtaining the LDA coefficients.This will give the importance scores associated with each feature.
lda_coefficients = np.exp(np.abs(lda.coef_))
lda_coefficients
```

```python
lda_coefficients = lda_coefficients.flatten() # any dimension to 1-D
lda_coefficients
```

```python
X.shape
```

```python
num_features = X.shape[1]
num_features
```

```python
feature_indices = np.arange(num_features) #96 - > 0 to 95
feature_indices
```

```python
feature_names = list(X.columns)
feature_names
```

```python
plt.figure(figsize=(20,18))

plt.bar(feature_indices, lda_coefficients)
plt.xticks(feature_indices, feature_names, rotation="vertical")
plt.xlabel("Feature")
plt.ylabel("Score")
plt.show()
```

```python
# Performing feature selection using LDA by using the absolute values of the
# LDA coefficients as a measure of feature importance.
```

```python
df_feature_score=pd.DataFrame({"Feature_names":feature_names,"feature_scores":lda_coefficients})
df_feature_score
```

```python
top_20_values = df_feature_score.nlargest(20, 'feature_scores')

## Selecting the top 20 features based on the feature importance
```

```python
top_20_values
```

```python
x_train = x_train[:, [17, 65, 7, 8, 40, 5, 20, 6, 21, 4, 14, 89, 64, 15, 18, 9, 16, 87,10, 12]]
```

```python
x_test = x_test[:,[17, 65, 7, 8, 40, 5, 20, 6, 21, 4, 14, 89, 64, 15, 18, 9, 16, 87,10, 12]]
```

```python
# Logistic Regression

from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()

lr.fit(x_train, y_train) #training
y_pred = lr.predict(x_test) #y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred, y_test)
```

```python
# Decision tree
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()

dt.fit(x_train, y_train)
y_pred = dt.predict(x_test)
```

```python
accuracy_score(y_pred, y_test)
```

```python
#KNN

from sklearn.neighbors import KNeighborsClassifier
knc= KNeighborsClassifier()

knc.fit(x_train,y_train)
y_pred=knc.predict(x_test)
```

```python
accuracy_score(y_pred, y_test)
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier

rfc=RandomForestClassifier()
rfc.fit(x_train, y_train)
y_pred=rfc.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
# XGBoost

import xgboost as xgb

xgbc = xgb.XGBClassifier()
xgbc.fit(x_train, y_train)
y_pred=xgbc.predict(x_test)
```

```python
accuracy_score(y_pred, y_test)
```
